##XGBoost

```python
import xgboost as xgb
from sklearn.metrics import ndcg_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import top_k_accuracy_score
```

```python
train, test = train_test_split(df, test_size=0.3, random_state=42)
```

```python
train_df = train.sort_values(by=['query_id']).reset_index(drop=True)
test_df = test.sort_values(by=['query_id']).reset_index(drop=True)
```

```python
X_train = train_df.drop(columns=['rank', 'query_id'])
y_train = train_df['rank']
qid_train = train_df['query_id']

X_test = test_df.drop(columns=['rank', 'query_id'])
y_test = test_df['rank']
qid_test = test_df['query_id']
```

```python
X_train
```

```python
ranker = xgb.XGBRanker(
    max_depth =  8,
    objective = 'rank:ndcg',
    eval_metric=["ndcg"],
    seed = 42
)
ranker.fit(
    X_train,
    y_train,
    qid=qid_train,
    verbose=True
)
```

```python
#метрика на трейне
y_pred_tr = ranker.predict(X_train)
ndcg_score([y_train], [y_pred_tr])
```

```python
#метрика на тесте
y_pred_test = ranker.predict(X_test)
ndcg_score([y_test], [y_pred_test])
```

##CatBoost

```python
#pip install catboost
import catboost as cat
```

```python
train_pool = cat.Pool(
    data=X_train,
    label=y_train,
    group_id=qid_train
)

test_pool = cat.Pool(
    data=X_test,
    label=y_test,
    group_id=qid_test
)
```

```python
ranker_ct = cat.CatBoostRanker(
    max_depth =  8,
    eval_metric="NDCG",
    random_seed = 42
)
ranker_ct.fit(
    train_pool,
    eval_set = test_pool
)
```

```python
#метрика на трейне
y_pred_tr = ranker_ct.predict(X_train)
ndcg_score([y_train], [y_pred_tr])
```

```python
#метрика на тесте
y_pred_test = ranker_ct.predict(X_test)
ndcg_score([y_test], [y_pred_test])
```
